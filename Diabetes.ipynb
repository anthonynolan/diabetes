```python
import pandas as pd
import numpy as np
dat = pd.read_csv('diabetes_data/diabetic_data.csv')
dat.set_index('encounter_id', inplace=True, verify_integrity=True)
dat.sort_values(['encounter_id'], inplace=True)
dat.reset_index(inplace=True)
```

```python
multiple_visit_patients = dat.groupby('patient_nbr').size().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
multiple_visit_patients[multiple_visit_patients>1].plot(kind='hist', bins=20)
plt.show()
```

```python
lower_limit = np.min(dat.encounter_id.values)
ticks = np.linspace(np.min(dat.encounter_id.values), np.max(dat.encounter_id.values), num=10)

if 'encounter_id' not in dat.columns:
    dat.reset_index(inplace=True)

for upper_limit in ticks[1:]:
    
    temp_dat = dat[(dat['encounter_id']<upper_limit )& (dat['encounter_id']>lower_limit)]
    lower_limit = upper_limit
    
    age_counts = temp_dat.age.value_counts().reset_index().sort_values('index')
    age_counts.columns = ['age', 'count']
    plt.figure()
    plt.title('upper limit {:,.0f}'.format(upper_limit))
    plt.barh(y = age_counts['age'], width = age_counts['count'])
    
```
